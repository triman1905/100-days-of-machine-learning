```
class Account:
    def __init__(self, name, balance, pin):
        self.name = name
        self.balance = balance
        self.pin = pin


class ATM:
    def __init__(self):
        pass

    def withdraw(self, account, pin, amount):
        if pin == account.pin:
            if amount <= account.balance:
                account.balance -= amount
                print(f"₹{amount} withdrawn successfully")
            else:
                print("Insufficient balance")
        else:
            print("Invalid Pin")

    def deposit(self, account, amount):
        account.balance += amount
        print(f"₹{amount} deposited successfully")

    def check_balance(self, account, pin):
        if pin == account.pin:
            print(f"Current balance: ₹{account.balance}")
        else:
            print("Invalid Pin")


# usage
acc = Account("Triman", 10000, "1234")
atm = ATM()

atm.withdraw(acc, "1234", 2000)
atm.deposit(acc, 500)
atm.check_balance(acc, "1234")
```

```
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def check_pass(self):
        if self.marks >= 40:
            print("Pass")
        else:
            print("Fail")


s1 = Student("Triman", 85)

s1.check_pass()

# dynamic attribute
s1.grade = "A"

print("Name:", s1.name)
print("Marks:", s1.marks)
print("Grade:", s1.grade)
```

```
class Bank:
    def __init__(self):
        self.__balance = 0

    def get_balance(self):
        return self.__balance

    def set_balance(self, new_value):
        if type(new_value) == int:
            self.__balance = new_value
        else:
            print("Invalid input! Balance must be integer")

    def deposit(self, amount):
        if type(amount) == int and amount > 0:
            self.__balance += amount
        else:
            print("Invalid deposit amount")


b1 = Bank()

print("Initial balance:", b1.get_balance())

b1.deposit(500)
print("After deposit:", b1.get_balance())

b1.set_balance(1000)
print("After setting balance:", b1.get_balance())

```

```
class Account:
  def __init__(self):
    self.__balance=0
  def deposit(self,amount):
    if amount>0:
      self.__balance+=amount
    else:
      print("Invalid amount")
  def withdraw(self,amount):
    if amount>0 and amount<=self.__balance:
      self.__balance-=amount
    else:
      print("Invalid amount")
  def transfer(self,to_account,amount):
    if amount>self.__balance:
      print("invalid")
    else:
      self.__balance -= amount
      to_account.deposit(amount)
a1 = Account()
a2 = Account()

a1.deposit(1000)
a1.transfer(a2, 500)
```

```
class Company:
    company_name = "Google"   # static variable

    def __init__(self, employee_name):
        self.employee_name = employee_name   # instance variable

    def display(self):
        print(f"{self.employee_name} works at {Company.company_name}")


e1 = Company("Triman")
e2 = Company("Rahul")

e1.display()
e2.display()
```

```
class Student:
    school_name = "DPS"   # static

    def __init__(self, name):
        self.name = name

    def display(self):
        print(f"{self.name} studies in {Student.school_name}")


s1 = Student("Triman")
s2 = Student("Aman")

s1.display()
s2.display()
```
